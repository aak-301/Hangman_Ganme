# My very first python project.It is a simple Hangaman Game Project.

```python
import os
stages=["""
        +---+
        |   |
        O   |
      /  \  |
       ||   |
      /  \  |
            |
===============
        
        ""","""
        +---+
        |   |
        O   |
      /  \  |
       ||   |
      /     |
            |
===============
        
        ""","""
        +---+
        |   |
        O   |
      /  \  |
       ||   |
            |
            |
===============
        
        ""","""
        +---+
        |   |
        O   |
      /  \  |
            |
            |
            |
===============
        
        ""","""
        +---+
        |   |
        O   |
            |
            |
            |
            |
===============
        
        ""","""
        +---+
        |   |
            |
            |
            |
            |
            |
===============
        
        """]
import random
context=["food","desert","ocean","city","state","country","car"]
context_index=random.randint(0,len(context)-1)
print(f"word is from group of {context[context_index]} names")
word_list=[["rice","chappati","daal","bhujia"],["choclate","kheer","sewai","gulabjamun"],["indian ocean","pacific ocean","atlantic ocean","arctic ocean"],["patna","varanasi","begusarai","paris"],["bihar","uttar pradesh","haryana","odisha"],["india","nepal","myanmar","japan"],["waganor","jaguar","nano","alto"]]
chosen_word=random.choice(word_list[context_index])
display=[]
lives=6
for i in range(0,len(chosen_word)):
    display.append("_")
print(display)
while display.count("_")!=0 and lives!=0:
    flag=0
    choose_word=input("guess a word: ")
    choose_word.lower()
    os.system("clear")
    for i in range(0,len(chosen_word)):
        if choose_word==chosen_word[i]:
            if display[i]=="_":
                display[i]=choose_word
                print(display)
                flag=1
                break
    if flag==0:
        print(stages[lives-1])
        lives-=1
if lives!=0:
    print("you won ðŸ˜Š")
else:
    print("You loose! ðŸ˜¥")

```
